```python
labels = pd.read_csv(r'C:\Users\e870859\Desktop\Data Science\7. Deep Learning Compititions\Tutorial -34 Head CT\Head CT Heamorrhage\labels.csv')[' hemorrhage'].tolist()
```

```python
import cv2
import glob
import numpy as np

X_data = []
files = glob.glob ("C:/Users/e870859/Desktop/Data Science/7. Deep Learning Compititions/Tutorial -34 Head CT/Head CT Heamorrhage/head_ct/head_ct/*.PNG")
for myFile in files:
    print(myFile)
    image = cv2.imread (myFile)
    img_resize = cv2.resize(image, (32, 32))
    X_data.append (img_resize)

print('X_data shape:', np.array(X_data).shape)
```

```python
# X = np.array(X_data).reshape(-1, 1)
```

```python
images_data = np.array(X_data)
```

```python
images_data.shape
```

```python
labels = np.array(labels)
```

```python
labels.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(images_data, labels, test_size = 0.2, random_state = 140)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
num_classes = len(np.unique(y_train))
num_classes
```

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train,num_classes)
y_test = to_categorical(y_test,num_classes)
```

```python
y_test
```

```python
print("X train shape: ",x_train.shape)
print("X test shape: ",x_test.shape)
print("y train shape: ",y_train.shape)
print("y test shape: ",y_test.shape)
```

```python
import tensorflow.keras as keras
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D,Flatten,Dense,Dropout
from keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator
```

```python
model = Sequential()
```

```python
x_train.shape[0:]
```

```python
model.add(Conv2D(filters=32, kernel_size=2, padding='same', activation='relu', input_shape=(32,32,3)))
model.add(MaxPool2D(pool_size=2))

model.add(Conv2D(filters=32, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=2))

model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=2))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(256,activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(64,activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(2,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='rmsprop', 
                  metrics=['accuracy'])
```

```python
from numpy.random import seed
seed(1337)
from tensorflow import set_random_seed
set_random_seed(1337)
```

```python
# data_gen = ImageDataGenerator(width_shift_range=0.1,
#                              height_shift_range=0.1,
#                              horizontal_flip=True)

data_gen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.0,
    zoom_range=0.1,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True)
```

```python
from keras.callbacks import ReduceLROnPlateau
reduce_learning_rate = ReduceLROnPlateau(monitor='loss',
                                         factor=0.1,
                                         patience=2,
                                         cooldown=2,
                                         min_lr=0.00001,
                                         verbose=1)

callbacks = [reduce_learning_rate]
```

```python
data_gen.fit(x_train)
```

```python
batch_size=40

history = model.fit_generator(data_gen.flow(x_train, y_train, batch_size=batch_size),
                    steps_per_epoch=x_train.shape[0] // batch_size,
                    epochs=30, verbose=2,callbacks=callbacks)
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['acc'])
# plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
# plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
print('Train accuracy of the model: ',history.history['acc'][-1])
```
